## Overview

## Input
- `Europe_Pharma_Job_Postings Combined Raw.xlsx` — combined dataset from 3 verified sources (30,516 rows).

##Inspection Steps are as follows
1. Check Raw Column Names
2. Check Post Date Type Consistency
3. Check for Missing Post Dates
4. Check for Duplicate Postings
5. Check Text Column Case Consistency
6. Check for Missing Categories
7. Check Salary Offered Format
8. Check Job Type Variants
9. Check Location Variants
10. Check Data Source and Category Breakdown

Install and Import Libraries

```python
!pip install pandas numpy -q

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("Libraries loaded")
```

Loading Raw Dataset

```python
df = pd.read_excel("/content/Europe_Pharma_Job_Postings Combined Raw.xlsx")
print(f"Loaded: {df.shape[0]:,} rows | {df.shape[1]} columns")
```

Checking Raw Column Names

```python
print("Raw column names:")
print(df.columns.tolist())
```

Checking the post_date column consistency

```python
print("post_date dtype:", df['Post_date'].dtype)
type_counts = df['Post_date'].apply(lambda x: type(x).__name__).value_counts()
print("Value types mixed within the column:", type_counts.to_dict())
```

Checking the missing post dates

```python
print("Rows with a blank/missing post_date:", df['Post_date'].isnull().sum())
```

Check for Duplicate postings

```python
dupe_count = df.duplicated(subset=['Job_title','Company_name','Location','Post_date','Job_description'], keep=False).sum()
full_dupes = df.duplicated(keep='first').sum()
print(f"Rows sitting in duplicate groups (title+company+location+date+description): {dupe_count:,}")
print(f"Fully identical rows (every column matches): {full_dupes:,}")
```

Checking consistency for text columns

```python
print("Inconsistent case examples: same job title but different casing:")
print(df['Job_title'].value_counts().head(10))
```

Check for Missing Categories

```python
print("Rows with missing category in raw data:", df['Category'].isnull().sum())
print(f"({df['Category'].isnull().sum() / len(df) * 100:.1f}% of total)")
```

Checking salary_offered format

```python
print("Top values in salary_offered (reason why text parsing + currency handling is needed):")
print(df['Salary_offered'].astype(str).value_counts().head(10))
```

Checking Job type variants

```python
print("Raw job_type: (showing inconsistent variants):")
print(df['Job_type'].value_counts())
```

Checking Location columns variants

```python
print("Raw location: showing combination of region/country granularity and case duplicates -")
print(df['Location'].value_counts())
```

Checking Data Source and Job category breakdown

```python
print("Data source breakdown:")
print(df['Data_Source'].value_counts())
print("\nCategory breakdown (top 10):")
print(df['Category'].value_counts().head(10))
```
