## Overview
This notebook helps to load combined pharmaceutical job postings dataset file and performs inspection to validate data quality
and structure before getting it ready for cleaning and analysis.

## Input
- `Combined_Pharma_Jobs.csv` — combined dataset from 9 European pharmaceutical job portals.

## Output
- Accurate row count, column names, data types
- Missing value report for each column
- Summary of statistics for key columns

Install and Import Libraries

```python
!pip install pandas numpy -q

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("Libraries loaded")
print(f"   pandas  version : {pd.__version__}")
print(f"   numpy   version : {np.__version__}")
```

```python
file_path = "/content/Combined_Pharma_Jobs.csv"
df = pd.read_csv(
    file_path,
    on_bad_lines='skip',
    engine='python'
)
df = df.drop(columns=['Unnamed: 8'], errors='ignore')

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
print(f"Loaded: {df.shape[0]:,} rows  |  {df.shape[1]} columns")
print(f"  Rows    : {df.shape[0]:,}")
print(f"  Columns : {df.shape[1]}")
```

```python

print("  SHAPE")
print(f"  Rows    : {df.shape[0]:,}")
print(f"  Columns : {df.shape[1]}")
print("  COLUMNS")
for i, col in enumerate(df.columns, 1):
    print(f"  {i}. {col}")
```

```python
print("  FIRST 5 ROWS")
print(df.head().to_string(index=False))
```

```python
print("  DATA TYPES")
print(f"\n  {'Column':<25} {'Type':<15}")
for col, dtype in df.dtypes.items():
    print(f"  {col:<25} {str(dtype):<15}")
```

```python
print("  MISSING VALUES")
print(f"\n  {'Column':<25} {'Missing':>8}   {'% Missing':>10}")
for col in df.columns:
    nulls = df[col].isnull().sum()
    pct   = (nulls / len(df)) * 100
    flag  = "Error" if pct > 10 else "Success"
    print(f"  {flag} {col:<23} {nulls:>8,}   {pct:>9.1f}%")
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print("Column names standardised:")
print(df.columns.tolist())
print("  QUICK SUMMARY")
print(f"  {'Total jobs':<25} : {len(df):>10,}")
print(f"  {'Unique job titles':<25} : {df['job_title'].nunique():>10,}")
print(f"  {'Unique locations':<25} : {df['location'].nunique():>10,}")
print(f"  {'Unique companies':<25} : {df['company_name'].nunique():>10,}")
print(f"  {'Unique categories':<25} : {df['category'].nunique():>10,}")
print(f"  {'Unique job types':<25} : {df['job_type'].nunique():>10,}")
print("  CATEGORY BREAKDOWN")
cat = df['category'].value_counts().reset_index()
cat.columns = ['Category', 'Count']
cat['Share %'] = (cat['Count'] / len(df) * 100).round(1)
print(cat.to_string(index=False))

print("  TOP 10 JOB TITLES")
titles = df['job_title'].value_counts().head(10).reset_index()
titles.columns = ['Job Title', 'Count']
print(titles.to_string(index=False))

print("  TOP 10 LOCATIONS")
locs = df['location'].value_counts().head(10).reset_index()
locs.columns = ['Location', 'Count']
print(locs.to_string(index=False))

print("  SALARY BREAKDOWN (top 10 values)")
sal = df['salary_offered'].value_counts().head(10).reset_index()
sal.columns = ['Salary Offered', 'Count']
print(sal.to_string(index=False))

print("  JOB TYPE BREAKDOWN")
jt = df['job_type'].value_counts().reset_index()
jt.columns = ['Job Type', 'Count']
jt['Share %'] = (jt['Count'] / len(df) * 100).round(1)
print(jt.to_string(index=False))
```
