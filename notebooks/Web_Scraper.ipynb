combining all csv files

```python
import csv
import os

csv_files = [
    '/content/Europharma_jobs.csv',
   '/content/Jobvector.de.csv',
   '/content/Pharmajobs.csv',
   '/content/emed_careers_eu 2.csv',
   '/content/englishjobs.csv',
   '/content/eurosciencejobs.csv',
   '/content/life-science.eu.csv',
   '/content/lifesciencenord.de.csv',
   '/content/pharmiweb.csv',
   '/content/pharmajobber.csv'
]

header = ['Category', 'Company Name', 'Job Description', 'Job Title', 'Job Type', 'Location', 'Post Date', 'Salary Offered']
all_rows = []

for filename in csv_files:
    if os.path.exists(filename):
        print(f"🔄 Adding {filename}...")
        with open(filename, 'r', encoding='utf-8') as f:
            reader = csv.reader(f)
            next(reader)
            for row in reader:
                if len(row) >= 7 and any(row):
                    row = row[:8] + [''] * (8 - len(row))
                    all_rows.append(row)
    else:
        print(f" Not found: {filename}")

out_file = 'ALL_Pharma_Jobs_NO_Dedupe.csv'
with open(out_file, 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(all_rows)

print(f" COMBINED {len(csv_files)} files → {len(all_rows)} TOTAL pharma jobs (duplicates KEPT)!")
print(f" Download '{out_file}' from Files tab")
print("\n Stats:")
print(f"Files: {len(csv_files)}")
print(f"Total rows: {len(all_rows)}")
```
